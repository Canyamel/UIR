```python
from typing import Any

class Word:
    def __init__(self, word, initial_form_word=None, part_speech=None, time=None, genus=None, number=None, case=None, animacy=None,
                comparison=None, brevity=None, representation=None, inclination=None, aspect=None, face=None, voice=None, additional_feature=None) -> None:
        self.word = word #слово
        self.initial_form_word = initial_form_word #начальная форма слова
        self.part_speech = part_speech #часть речи
        self.time = time #время
        self.genus = genus #род
        self.number = number #число
        self.case = case #падеж
        self.animacy = animacy #одушевлённость
        self.comparison = comparison #сравнение
        self.brevity = brevity #краткость
        self.representation = representation #репрезентация
        self.inclination = inclination #наклонение
        self.aspect = aspect #вид
        self.face = face #лицо
        self.voice = voice #залог
        self.additional_feature = additional_feature #дополнительные характеристики

    def __repr__(self):
        return f"{self.word}"

class NodeWord:
    def __init__(self, word: Word, index: int) -> None:
        self.word = word
        self.index = index

    def __repr__(self):
        return f"{self.word}"

class Edge:
    def __init__(self, start, end, mark=None) -> None:
        self.start = start
        self.end = end
        self.mark = mark

    def __repr__(self):
        return f"({self.start}, {self.end})"

class SyntacticTree:
    def __init__(self, list_NodeWord, list_Edge, root: int) -> None:
        self.list_NodeWord = list_NodeWord
        self.list_Edge = list_Edge
        self.root = root

from itertools import combinations
import copy

def find_duplicates(input_list: list):
    buf_list = list()
    for item in input_list:
        if item not in buf_list:
            buf_list.append(item)
        else:
            return True
    return False

def get_single_level_list(input_list: list):
    new_list = list()
    for x in input_list:
        if type(x) == list:
            for item in x:
                new_list.append(item)
        else:
            new_list.append(x)

    return new_list

def get_combinations(numbers):
    combinations_list = list()
    for i in range(2, len(numbers) + 1):
        for combination in combinations(numbers, i):
            combinations_list.append(get_single_level_list(list(combination)))

    res_combinations_list = list()
    for comb in combinations_list:
        if find_duplicates(comb) == False:
            res_combinations_list.append(comb)

    return res_combinations_list

def get_elements(tree: SyntacticTree, owner: int):
    servants = list()
    elements = list([owner])

    for Edge in tree.list_Edge:
        if Edge.start == owner:
            servants.append(Edge.end)

    if len(servants) > 0:
        for serv in servants:
            elements.extend(get_elements(tree, serv))

        combinations = get_combinations(elements)
        result = list()
        for comb in combinations:
            if owner in comb:
                result.append(comb)
        return result + [owner]

    return [owner]

def get_subtrees(tree: SyntacticTree, owner: int):
    subtrees = list()
    servants = list()
    elements = list([owner])

    for Edge in tree.list_Edge:
        if Edge.start == owner:
            servants.append(Edge.end)

    for serv in servants:
        elements.extend(get_elements(tree, serv))

    res_subtrees = list()
    subtrees = get_combinations(elements)
    for subtree in subtrees:
        if owner in subtree and len(subtree) < len(tree.list_NodeWord):
            subtree.sort()
            res_subtrees.append(subtree)

    return res_subtrees

def get_all_subtrees(tree: SyntacticTree):
    subtrees = list()
    for index in range(len(tree.list_NodeWord)):
        subtrees.extend(get_subtrees(tree, index))

    return subtrees

list_NodeWord = [
    NodeWord(Word("Красивый"), 0),
    NodeWord(Word("мальчик"), 1),
    NodeWord(Word("пошёл"), 2),
    NodeWord(Word("в"), 3),
    NodeWord(Word("парк"), 4),
    #NodeWord(Word("test1"), 5),
    #NodeWord(Word("test2"), 6),
    #NodeWord(Word("test3"), 7),
    #NodeWord(Word("test4"), 8)
]

list_Edge = [
    Edge(2, 1),
    Edge(2, 3),
    Edge(3, 4),
    Edge(1, 0),
    #Edge(3, 5),
    #Edge(3, 6),
    #Edge(0, 7),
    #Edge(0, 8)
]

tree = SyntacticTree(list_NodeWord, list_Edge, 2)

get_all_subtrees(tree)
#get_subtrees(tree, 2)
```
